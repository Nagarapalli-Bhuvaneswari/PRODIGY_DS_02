```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\Bhuva\Downloads\Task 2\Data_Train.csv')  
```

```python
print(df.isnull().sum()) 
```

```python
df.dropna(subset=['Price'], inplace=True) 
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y')
```

```python
df['Journey_Day'] = df['Date_of_Journey'].dt.day
df['Journey_Month'] = df['Date_of_Journey'].dt.month
df['Journey_Year'] = df['Date_of_Journey'].dt.year
```

```python
df['Dep_Hour'] = df['Dep_Time'].dt.hour
df['Dep_Minute'] = df['Dep_Time'].dt.minute
df['Arrival_Hour'] = df['Arrival_Time'].dt.hour
df['Arrival_Minute'] = df['Arrival_Time'].dt.minute
```

```python
df['Duration'] = df['Duration'].str.replace('h', '*60').str.replace('m', '').str.replace(' ', '').apply(eval)
df['Duration'] = df['Duration'].apply(lambda x: sum(int(i) for i in str(x).split('*60')))
```

```python
stops = {'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4}
df['Total_Stops'] = df['Total_Stops'].map(stops)
```

```python
df.drop(['Date_of_Journey', 'Dep_Time', 'Arrival_Time'], axis=1, inplace=True)
```

```python
# EXPLORATORY DATA ANALYSIS
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Airline', data=df)
plt.xticks(rotation=90)
plt.title('Count of Flights for Each Airline')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Source', data=df)
plt.title('Count of Flights from Each Source City')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Destination', data=df)
plt.title('Count of Flights to Each Destination City')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Airline', y='Price', data=df)
plt.xticks(rotation=90)
plt.title('Price Distribution for Each Airline')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Total_Stops', y='Price', data=df)
plt.title('Price Distribution Based on Number of Stops')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Duration', y='Price', hue='Airline', data=df)
plt.title('Relationship Between Duration and Price')
plt.show()
```
